## 感知机

$$\binom{n+1}{2k} $$

感知机器原理数学描述 

$$
y = 
\begin{cases}
0 \quad (w_1x_1 + w_2x_2) \leq \theta \\
1 \quad (w_1x_1 + w_2x_2) > \theta
\end{cases}
\tag{1}
$$

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1 * w1 + x2 * w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1


print(AND(0.8, 0.7))
```

$$
y = 
\begin{cases}
0 \quad (b + w_1x_1 + w_2x_2) \leq 0 \\
1 \quad (b + w_1x_1 + w_2x_2) > 0
\end{cases}
\tag{1}
$$

b 偏置，w<sub>1</sub>, w<sub>2</sub> 权重
